```python
# Importing NumPy
import numpy as np

# 11. Mathematical Functions
array = np.array([1, 2, 3, 4, 5])

# Applying mathematical functions
squared = np.square(array)
sqrt = np.sqrt(array)
sin_values = np.sin(array)
log_values = np.log(array)

print("Original Array:", array)
print("Squared:", squared)
print("Square Root:", sqrt)
print("Sine Values:", sin_values)
print("Logarithm Values:", log_values)

print("\n")  # Add space between sections

# 12. Random Number Generation
# Random integers
random_ints = np.random.randint(1, 10, size=(3, 3))

# Random floats
random_floats = np.random.rand(3, 3)

# Normal distribution
random_normal = np.random.normal(0, 1, size=(3, 3))

print("Random Integers:\n", random_ints)
print("\nRandom Floats:\n", random_floats)
print("\nRandom Normal Distribution:\n", random_normal)

print("\n")  # Add space between sections

# 13. Sorting and Unique Values
array = np.array([5, 2, 9, 1, 5, 6, 7, 8, 2])

# Sorting the array
sorted_array = np.sort(array)

# Finding unique values
unique_values = np.unique(array)

print("Original Array:", array)
print("Sorted Array:", sorted_array)
print("Unique Values:", unique_values)

print("\n")  # Add space between sections

# 14. Aggregations
array = np.array([1, 2, 3, 4, 5])

# Aggregate functions
total_sum = np.sum(array)
average = np.mean(array)
max_value = np.max(array)
min_value = np.min(array)

print("Original Array:", array)
print("Sum:", total_sum)
print("Mean:", average)
print("Max:", max_value)
print("Min:", min_value)

print("\n")  # Add space between sections

# 15. Linear Algebra
matrix = np.array([[1, 2], [3, 4]])

# Determinant
determinant = np.linalg.det(matrix)

# Inverse
inverse_matrix = np.linalg.inv(matrix)

print("Original Matrix:\n", matrix)
print("Determinant:", determinant)
print("Inverse Matrix:\n", inverse_matrix)
```
